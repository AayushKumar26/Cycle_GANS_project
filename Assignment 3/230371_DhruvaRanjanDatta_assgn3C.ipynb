```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.python.keras import layers
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import f1_score, accuracy_score
```

```python
data = pd.read_csv("/content/drive/MyDrive/Copy of Train.csv")
data
```

```python
data.isna().sum()
```

```python
data['MatchWinner'].value_counts()
```

```python
categories = ['Team1', 'Team2', 'Stadium', 'HostCountry', 'Team1_Venue', 'Team2_Venue', 'Team1_Innings', 'Team2_Innings', 'MonthOfMatch']
X = data[categories]
y = data[['MatchWinner']]
```

```python
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(X[['Team1', 'Team2', 'Stadium', 'HostCountry', 'Team1_Venue', 'Team2_Venue', 'Team1_Innings', 'Team2_Innings', 'MonthOfMatch']])
```

```python
X_transformed = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(['Team1', 'Team2', 'Stadium', 'HostCountry', 'Team1_Venue', 'Team2_Venue', 'Team1_Innings', 'Team2_Innings', 'MonthOfMatch']))
X_transformed
```

```python
X_train = X_transformed[:2108]
y_train = y[:2108]
X_test = X_transformed[2108:]
y_test = y[2108:]
```

```python
model = tf.keras.Sequential([
    layers.Flatten(input_shape=(2108, 218)),
    layers.Dense(700, activation="relu"),
    layers.Dense(350, activation="relu"),
    layers.Dense(16, activation="softmax")
])
```

```python
model.compile(optimizer="adam", loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
probs = model.predict(X_test)
y_pred = np.argmax(probs, axis=1)
y_pred
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(test_acc)
```

```python
accuracy_score(y_test, y_pred)
```
