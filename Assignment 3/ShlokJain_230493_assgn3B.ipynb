# Part B (Neural Network from Scratch)

You need to implement a neural network from scratch .This is a multiclass classification problem. No. of hidden layers depends on you but should be atleast 2.Remember to use activation function. You can add any other function of your choice.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
#train test split
X_train = X[:125]
X_test = X[125:]
y_train = y[:125]
y_test = y[125:]
```

```python
print(X_train.T.shape,y.shape)
```

```python
#convert y to one-hot encoding
# y = np.eye(3)[y]
# y = y.T
# X = X.T
y_train = np.eye(3)[y_train]
y_test = np.eye(3)[y_test]
X_train = X_train.T
X_test = X_test.T
y_train = y_train.T
y_test = y_test.T
```

```python
y.shape
```

```python
def parameters(n1,n2):
  # define the parameters of your nn initially using random lib.
  n = X.shape[0]
  p = {}
  a = {}
  p["w1"] = np.random.rand(n1,4)
  p["w2"] = np.random.rand(n2,n1)
  p["w3"] = np.random.rand(3,n2)
  p["b1"] = np.random.rand(n1,1)
  p["b2"] = np.random.rand(n2,1)
  p["b3"] = np.random.rand(3,1)
  a["a1"] = np.zeros((n,n1))
  a["a2"] = np.zeros((n,n2))
  a["a3"] = np.zeros((n,3))
  return p,a
```

```python
#activation functions
def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return np.maximum(0,x)

def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
def forward(X,p,a):
    # function for forward propagation
    # for i in range(150):
    #     a["a1"][i] = relu(np.dot(p["w1"],X[i].T)+p["b1"])
    #     a["a2"][i] = relu(np.dot(p["w2"],a["a1"][i].T)+p["b2"])
    #     a["a3"][i] = softmax(np.dot(p["w3"],a["a2"][i].T)+p["b3"])
    a1 = relu(np.dot(p["w1"],X)+p["b1"])
    a2 = relu(np.dot(p["w2"],a1)+p["b2"])
    a3 = softmax(np.dot(p["w3"],a2)+p["b3"])
    a["a1"] = a1
    a["a2"] = a2
    a["a3"] = a3
    return a
```

```python
def cost_funct(y,a3):
    m = y.shape[1]
    cost = -np.sum(y*np.log(a3))/m
    return cost
```

```python
# use Gradient descent as of now as an optimizer
```

```python
def backward(X,y,a3,p,a):
    m = X.shape[1]
    dz3 = a3 - y
    dw3 = np.dot(dz3,a["a2"].T)/m
    db3 = np.sum(dz3,axis=1,keepdims=True)/m
    dz2 = np.dot(p["w3"].T,dz3) * (a["a2"] > 0)
    dw2 = np.dot(dz2,a["a1"].T)/m
    db2 = np.sum(dz2,axis=1,keepdims=True)/m
    dz1 = np.dot(p["w2"].T,dz2) * (a["a1"] > 0)
    dw1 = np.dot(dz1,X.T)/m
    db1 = np.sum(dz1,axis=1,keepdims=True)/m
    return dw1,db1,dw2,db2,dw3,db3
```

```python
def update_parameters(p,dw1,db1,dw2,db2,dw3,db3,lr):
#FUNCTION TO UPDATE PARAMETERS USING GD
    p["w1"] = p["w1"] - lr*dw1
    p["w2"] = p["w2"] - lr*dw2
    p["w3"] = p["w3"] - lr*dw3
    p["b1"] = p["b1"] - lr*db1
    p["b2"] = p["b2"] - lr*db2
    p["b3"] = p["b3"] - lr*db3
    return p
```

```python
p,a = parameters(10,10)
def model(X,y,epochs,lr,p,a):
    #function to train and build the whole model
    for i in range(epochs):
        a = forward(X,p,a)
        cost = cost_funct(y,a["a3"])
        dw1,db1,dw2,db2,dw3,db3 = backward(X,y,a["a3"],p,a)
        p = update_parameters(p,dw1,db1,dw2,db2,dw3,db3,lr)
        print("Epoch: ",i,"Cost: ",cost)
    return p
  #function to train and build the whole model
p = model(X_train,y_train,2000,0.001,p,a)
```

```python
#write down the predictions and the f1 score finally
def predict(X,p):
    a1 = relu(np.dot(p["w1"],X)+p["b1"])
    a2 = relu(np.dot(p["w2"],a1)+p["b2"])
    a3 = softmax(np.dot(p["w3"],a2)+p["b3"])
    return a3

pred = predict(X_test,p).T
# pred = np.argmax(pred,axis=0)
# y = np.argmax(y,axis=0)
pred = np.argmax(pred,axis=1)
y_new = np.argmax(y_test,axis=0)
from sklearn.metrics import f1_score
f1_score(y_new,pred,average='weighted')

```
